```python
import tensorflow as tf
from transformers import MarianTokenizer, TFAutoModelForSeq2SeqLM
from sklearn.model_selection import train_test_split
from nltk.translate.bleu_score import corpus_bleu, sentence_bleu
import pandas as pd
```

```python
# Load and preprocess train and test data
train_data = pd.read_csv(dir_path_train)  # Modify with the path to your train CSV file
test_data = pd.read_csv(dir_path_test)    # Modify with the path to your test CSV file

english_train = train_data['en'].tolist()
french_train = train_data['fr'].tolist()
english_test = test_data['en'].tolist()
french_test = test_data['fr'].tolist()
```

```python
# Split the data into training and test sets
#train_data, test_data = train_test_split(test_data, test_size=0.2)
train_data = train_data.dropna(axis=0, how="any", subset=None, inplace=False)
test_data = test_data.dropna(axis=0, how="any", subset=None, inplace=False)
```

```python
# Define your model and tokenizer

!pip install sentencepiece
from transformers import MarianTokenizer
model_name = "Helsinki-NLP/opus-mt-en-fr"
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = TFAutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
# Translate English sentences to French
translated_sentences = []

# for sentence in english_test:
sentence=english_test[0]
inputs = tokenizer.encode(sentence, return_tensors="tf")
translated = model.generate(inputs, max_length=60, num_beams=4, no_repeat_ngram_size=1, top_k=50, eos_token_id=tokenizer.eos_token_id)
translated_text = tokenizer.decode(translated[0], skip_special_tokens=True)
translated_sentences.append(translated_text)
print(sentence)
print("\n",translated_text)
print()
print(french_test[0])
```
